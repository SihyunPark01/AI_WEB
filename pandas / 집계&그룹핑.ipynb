```python
#  그룹핑 / 집계하기
import pandas as pd
path = 'https://raw.githubusercontent.com/blackdew/ml-tensorflow/master/data/csv/iris.csv'
df = pd.read_csv(path)
df.head()
```

```python
print("sum()")
print(df[['꽃잎길이', '꽃잎폭']].sum())

print("sum()")
print(df[['꽃잎길이', '꽃잎폭']].mean())

print("sum()")
print(df[['꽃잎길이', '꽃잎폭']].std())

print("sum()")
print(df[['꽃잎길이', '꽃잎폭']].median())

print("sum()")
print(df[['꽃잎길이', '꽃잎폭']].min())

```

```python
# 합치기
import pandas as pd
df1 = pd.DataFrame([[1,2,3,4],[4,5,6,7],[7,8,9,10]])
df2 = pd.DataFrame([[1,2,3,5],[4,5,6,7],[7,8,9,10]])
print(df1.shape, df2.shape)
```

```python
df_concat = pd.concat([df1, df2], axis=1)  # 우측으로 합치기, 축을 1번째 방향으로 합쳐라
print(df_concat.shape)
df_concat
```

```python
df_concat.columns = [0,1,2,3,4,5,6,7]
df_concat
```

```python
df_concat = pd.concat([df1, df2], axis=0) # 아래로 합치기, 축을 0번째 방향으로 합쳐라
print(df_concat.shape)
df_concat
```

```python
df_concat.sum(axis=0)
```

```python
df_concat.mean(axis=0)
df_concat.mean(axis=0).shape
```

```python
df_concat = pd.concat([df1, df2], axis=0)
print(df_concat.shape)
df_concat
```

```python
df_concat = df_concat.reset_index(drop=True)
df_concat
```

```python
# 합치기 정리
# concat - dataframe을 붙인다
# axis 0 : row 방향으로 합침
# axis 1 : column 방향으로 합침
# reset_index - index를 재설정함
# column 할당 - column 이름을 새로 지정함
```
