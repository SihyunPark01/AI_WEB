```python
# 상관관계 : 변수 간의 관련성 확인
# 수치형 변수 : 수치형 변수
# 범주형 변수 : 범주형 변수
# 범주형 변수 : 수치형 변수
```

```python
# 수치형 변수 : 수치형 변수
# 상관계수 ; -1~1사이의 수
import seaborn as sns
import pandas as pd
df = sns.load_dataset('tips')

# scatterplot
sns.scatterplot(x='total_bill', y='tip', data=df)

# lmplot : scatter plot + 회귀선
sns.lmplot(data=df, x='total_bill', y='tip')
```

```python
# tip은 total_bill과 얼마나 관련이 있을까? 상관계수로 파악가능!
df.corr() #1에 가까울수록 양의 상관관계가 높고, 0에 가까울수록 상관관계가 없다. -1에 가까울수록 음의 상관관계가 높다
# 직선의 기울기와는 관련이 없음
# 선형성이 아닌 패턴은 잡아내지 못한다는 한계가 있음
```

```python
# 범주형 변수 : 범주형 변수
# cross table, contingency table
# grouped bar graph
# sex, smoker, 
# sex와 smoker은 서로 관련이 있을까?

# 절대빈도
pd.crosstab(df['sex'], df['smoker'], margins=True)
```

```python
# 상대빈도
pd.crosstab(df['sex'], df['smoker'], margins=True, normalize='all')
```

```python
# 행방향 상대빈도
pd.crosstab(df['sex'], df['smoker'], margins=True, normalize='index')
```

```python
# 열방향 상대빈도
pd.crosstab(df['sex'], df['smoker'], margins=True, normalize='columns')
```
